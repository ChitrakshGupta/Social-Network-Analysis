## program to calculate degree distribution


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = int(input("no of nodes"))
print("row")
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
matrix=np.array(matrix)
```

```python
import matplotlib.pyplot as plt

def plot_degree_distribution(graph):
    degrees = [sum(node) for node in graph]
    labels = [f"u{i+1}" for i in range(len(graph))]

    # Plotting as a line chart
    plt.plot(labels, degrees, color='red', marker='o', linestyle='-')
    plt.title('Degree Distribution')
    plt.xlabel('Nodes')
    plt.ylabel('Degree')
    plt.xticks(rotation=45)
    plt.grid(True)  
    plt.show()
plot_degree_distribution(matrix)
```
